```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'melb_data.csv')
```

```python
data.shape
data.columns
```

```python
data.replace(['h', 'u', 't']
                ,['Bungalow', 'Apartment', 'Townhouse'],inplace=True)
```

```python
data = data.drop(['CouncilArea', 'Method', 'Date'], axis=1)
```

```python
data = data.rename(index=str, columns={"Bedroom2": "Bedroom", "Landsize": "OpenArea", "Distance": "Dist. to city"})
```

```python
data.shape
```

```python
data.isna().sum() 
```

```python
is_year = data['YearBuilt'].notnull()
data = data[is_year]
```

```python
data['Age'] = 2019 - data['YearBuilt']
```

```python
data = data.drop(['YearBuilt'], axis = 1)
```

```python
data
```

```python
def family_size(row):
    if row['Rooms'] == 1:
        return np.random.randint(1,3)
    elif row['Rooms'] == 2:
        return np.random.randint(1,5)
    elif row['Rooms'] == 3:
        return np.random.randint(2,7)
    elif row['Rooms'] == 4:
        return np.random.randint(2,7)
    elif row['Rooms'] == 5:
        return np.random.randint(2,7)
    else:
        return np.random.randint(4,9)
```

```python
data['family_size'] = data.apply(lambda row: family_size(row),axis=1)
```

```python
data
```

```python
def children(row):
    if row['family_size'] == 1:
        return 0
    elif row['family_size'] == 2:
        return np.random.choice(2, p=[0.9, 0.1])
    elif row['family_size'] == 3:
        return np.random.choice(3, p=[0.1, 0.7, 0.2])
    elif row['family_size'] == 4:
        return np.random.choice(3, p=[0.1, 0.2, 0.7])
    elif row['family_size'] == 5:
        return np.random.choice(5, p=[0.05, 0.2, 0.3, 0.4, 0.05])
    else:
        return np.random.choice(6, p=[0.01, 0.03, 0.5, 0.4, 0.05, 0.01])
```

```python
data['children'] = data.apply(lambda row: children(row),axis=1)
```

```python
data
```

```python
sub_data1 = data.iloc[:, 0]
suburb = sub_data1.value_counts().keys().tolist()
values = sub_data1.value_counts().tolist()
```

```python
plt.gcf().set_size_inches(25,15)
plt.xlabel('Suburb')
plt.ylabel('Size')
plt.title('Children | Family')
def subcategorybar(X, vals, width=0.8):
    n = len(vals)
    _X = np.arange(len(X))
    plt.bar(_X, vals[0], width=width/n, color = '#F51F05')
    plt.bar(_X, vals[1], width=width/n, color = '#05F579')
    plt.xticks(_X, X)

subcategorybar(suburb, [family_size, children])
label = ['Family', 'Children']
plt.legend(labels = label)
plt.gcf().autofmt_xdate()
plt.show()
```
